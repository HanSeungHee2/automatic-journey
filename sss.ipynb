```python
# poisson distribution
%matplotlib inline
import matplotlib.pyplot as plt
import pylab 
import numpy as np 
import sympy
import math
import matplotlib.ticker as ticker
k   = np.arange(0,10,1)


exp = 1
foo = []
for i in k:
    y = ((math.exp(-1*exp)*(exp**i))/(math.factorial(i)))
    foo.append(y)

plt.plot(k,foo)
ax=plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

plt.show()
```
